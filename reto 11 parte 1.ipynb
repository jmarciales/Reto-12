# Problema 1
Desarrollar un algoritmo que imprima de manera ascendente los valores (todos del mismo tipo) de un diccionario.

```python
def imprimir_valores_ascendentes(diccionario):


    # Extraer los valores del diccionario
    valores = list(diccionario.values())

    # Ordenar los valores de manera ascendente
    valores.sort()

    # Imprimir valores ordenados
    print("Valores ordenados de manera ascendente:")
    for valor in valores:
        print(valor)

# Ejemplo de uso del Problema 1:
print()
calificaciones = {
    "Matemáticas": 85,
    "Español": 92,
    "Historia": 78,
    "Ciencias": 90,
    "Inglés": 88
}
print("Diccionario original de calificaciones:")
print(calificaciones)
imprimir_valores_ascendentes(calificaciones)

estudiantes_edades = {
    "Carlos": 19,
    "Ana": 22,
    "Miguel": 18,
    "Laura": 20,
    "José": 21
}
print("\nDiccionario original de edades:")
print(estudiantes_edades)
imprimir_valores_ascendentes(estudiantes_edades)

```

# Problema 2
Desarrollar una funci�on que reciba dos diccionarios como par�ametros y los mezcle, es decir, que se construya un nuevo diccionario con las llaves de los dos diccionarios; si hay una clave repetida en ambos diccionarios, se debe asignar el valor

```python
def mezclar_diccionarios(dict1, dict2):
    # Crear un nuevo diccionario con los valores del segundo diccionario
    resultado = dict2.copy()

    # Actualizar con los valores del primer diccionario (estos tendrán prioridad)
    resultado.update(dict1)

    return resultado


print()

info_personal = {
    "nombre": "Laura",
    "edad": 20,
    "email": "laura@ejemplo.com",
    "ciudad": "Bogotá"
}

info_academica = {
    "carrera": "Ingeniería",
    "semestre": 4,
    "promedio": 8.7,
    "ciudad": "Medellín"  # Clave repetida con diferente valor
}

print("Primer diccionario (info_personal):")
print(info_personal)
print("\nSegundo diccionario (info_academica):")
print(info_academica)

# Mezclando los diccionarios
info_completa = mezclar_diccionarios(info_personal, info_academica)
print("\nDiccionario mezclado (info_completa):")
print(info_completa)
print("Nota: La clave 'ciudad' tiene el valor del primer diccionario ('Bogotá')")

# Ejemplo 2: Configuración de aplicación
config_predeterminada = {
    "tema": "claro",
    "idioma": "español",
    "notificaciones": True,
    "volumen": 80
}

config_usuario = {
    "tema": "oscuro",
    "sincronizacion": True,
    "volumen": 50
}

print("\nPrimer diccionario (config_predeterminada):")
print(config_predeterminada)
print("\nSegundo diccionario (config_usuario):")
print(config_usuario)

# Mezclando las configuraciones
config_final = mezclar_diccionarios(config_usuario, config_predeterminada)
print("\nDiccionario mezclado (config_final):")
print(config_final)
print("Nota: Las claves 'tema' y 'volumen' tienen los valores del primer diccionario (config_usuario)")
```

Cree un programa que lea de un archivo con dicho JSON y:

Imprima los nombres completos (nombre y apellidos) de las personas que practican el deporte ingresado por el usuario.
Imprima los nombres completos (nombre y apellidos) de las personas que est�en en un rango de edades dado por el usuario.

```python
import json

def cargar_json_desde_archivo(nombre_archivo):

    try:
        with open(nombre_archivo, 'r', encoding='utf-8') as archivo:
            return json.load(archivo)
    except FileNotFoundError:
        print(f"Error: No se encontró el archivo {nombre_archivo}")
        return None
    except json.JSONDecodeError:
        print(f"Error: El archivo {nombre_archivo} no contiene JSON válido")
        return None

def personas_por_deporte(datos, deporte_buscado):

    resultado = []

    # Recorrer cada persona en el diccionario
    for persona_id, datos_persona in datos.items():
        # Verificar si la persona practica el deporte buscado
        if deporte_buscado in datos_persona["deportes"]:
            nombre_completo = f"{datos_persona['nombres']} {datos_persona['apellidos']}"
            resultado.append(nombre_completo)

    return resultado

def personas_por_rango_edad(datos, edad_min, edad_max):

    resultado = []

    # Recorrer cada persona en el diccionario
    for persona_id, datos_persona in datos.items():
        # Verificar si la edad de la persona está en el rango
        if edad_min <= datos_persona["edad"] <= edad_max:
            nombre_completo = f"{datos_persona['nombres']} {datos_persona['apellidos']}"
            resultado.append(nombre_completo)

    return resultado

# Ejemplo de uso del Problema 3:


# Datos de ejemplo para el problema 3
datos_json = {
    "jadiazcoronado": {
        "nombres": "Juan Antonio",
        "apellidos": "Díaz Coronado",
        "edad": 19,
        "colombiano": True,
        "deportes": ["Fútbol", "Ajedrez", "Gimnasia"]
    },
    "dmlunasol": {
        "nombres": "Dorotea Maritza",
        "apellidos": "Luna Sol",
        "edad": 25,
        "colombiano": False,
        "deportes": ["Baloncesto", "Ajedrez", "Gimnasia"]
    },
    "pgomez": {
        "nombres": "Pedro",
        "apellidos": "Gómez López",
        "edad": 19,
        "colombiano": True,
        "deportes": ["Natación", "Tenis"]
    },
    "mrodriguez": {
        "nombres": "María",
        "apellidos": "Rodríguez Martínez",
        "edad": 22,
        "colombiano": True,
        "deportes": ["Fútbol", "Baloncesto"]
    }
}

# Guardamos los datos en un archivo JSON para simular la lectura desde archivo
nombre_archivo = "personas.json"
with open(nombre_archivo, "w", encoding="utf-8") as archivo:
    json.dump(datos_json, archivo, ensure_ascii=False, indent=4)

print(f"Archivo '{nombre_archivo}' creado con los datos de ejemplo")

# Simulamos la carga de datos desde el archivo
datos = cargar_json_desde_archivo(nombre_archivo)


deporte_buscado = "Fútbol"
personas = personas_por_deporte(datos, deporte_buscado)
print(f"\nPersonas que practican {deporte_buscado}:")
for persona in personas:
    print(f"- {persona}")

# Ejemplo: Buscar personas por rango de edad
edad_min, edad_max = 18, 20
personas = personas_por_rango_edad(datos, edad_min, edad_max)
print(f"\nPersonas con edades entre {edad_min} y {edad_max} años:")
for persona in personas:
    print(f"- {persona}")

# Función principal para ejecutar el programa interactivamente
def main():
    datos = datos_json  # Usamos los datos de ejemplo

    while True:
        print("\n--- Menú ---")
        print("1. Buscar personas por deporte")
        print("2. Buscar personas por rango de edad")
        print("3. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            deporte = input("Ingrese el deporte a buscar: ")
            personas = personas_por_deporte(datos, deporte)
            if personas:
                print(f"\nPersonas que practican {deporte}:")
                for persona in personas:
                    print(f"- {persona}")
            else:
                print(f"No se encontraron personas que practiquen {deporte}")

        elif opcion == "2":
            try:
                edad_min = int(input("Ingrese la edad mínima: "))
                edad_max = int(input("Ingrese la edad máxima: "))
                personas = personas_por_rango_edad(datos, edad_min, edad_max)
                if personas:
                    print(f"\nPersonas con edades entre {edad_min} y {edad_max} años:")
                    for persona in personas:
                        print(f"- {persona}")
                else:
                    print(f"No se encontraron personas en el rango de edad {edad_min}-{edad_max}")
            except ValueError:
                print("Error: Ingrese valores numéricos para las edades")

        elif opcion == "3":
            print("¡Hasta luego!")
            break

        else:
            print("Opción no válida. Intente de nuevo.")


if __name__ == "__main__":
   main()

```
