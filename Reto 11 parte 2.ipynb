```python
import requests
import json
from pprint import pprint

def conectar_api(url, params=None, headers=None):

    try:
        # Realizamos la solicitud GET a la API
        response = requests.get(url, params=params, headers=headers)

        # Verificamos si la solicitud fue exitosa (código 200)
        response.raise_for_status()

        # Convertimos la respuesta a formato JSON
        return response.json()

    except requests.exceptions.RequestException as e:
        print(f"Error al conectar con la API {url}: {e}")
        return None

def extraer_pares_clave_valor(datos, prefijo=""):

    if isinstance(datos, dict):
        # Si es un diccionario, iteramos por sus claves y valores
        for clave, valor in datos.items():
            # Si el valor es un diccionario o una lista, lo procesamos recursivamente
            if isinstance(valor, (dict, list)):
                print(f"{prefijo}{clave}: (objeto complejo)")
                extraer_pares_clave_valor(valor, prefijo + "  ")
            else:
                # Si es un valor simple, lo imprimimos
                print(f"{prefijo}{clave}: {valor}")

    elif isinstance(datos, list):
        # Si es una lista, procesamos cada elemento
        if datos and len(datos) > 0:
            print(f"{prefijo}Lista con {len(datos)} elementos")
            # Mostramos solo el primer elemento como ejemplo
            if len(datos) > 0:
                print(f"{prefijo}Primer elemento:")
                extraer_pares_clave_valor(datos[0], prefijo + "  ")

def api_pokemon():
    """Conecta a la API de Pokémon y obtiene datos de un Pokémon"""
    print("\n===== API 1: PokeAPI =====")
    url = "https://pokeapi.co/api/v2/pokemon/ditto"

    datos = conectar_api(url)
    if datos:
        print("Conexión exitosa con PokeAPI!")
        print("\nEstructura JSON completa (resumida):")
        pprint(datos, depth=1)  # Mostramos el JSON resumido

        print("\nExtracción de pares clave-valor principales:")
        # Extraemos algunos datos específicos que nos interesan
        print(f"Nombre: {datos.get('name')}")
        print(f"ID: {datos.get('id')}")
        print(f"Altura: {datos.get('height')}")
        print(f"Peso: {datos.get('weight')}")

        print("\nHabilidades:")
        for habilidad in datos.get('abilities', []):
            print(f" - {habilidad.get('ability', {}).get('name')}")

        print("\nTodos los pares clave-valor (muestra parcial):")
        # Mostramos una extracción parcial para no saturar la salida
        extraer_pares_clave_valor({'name': datos.get('name'),
                                  'id': datos.get('id'),
                                  'abilities': datos.get('abilities')[:2]})
    else:
        print("No se pudieron obtener datos de la API de Pokémon")

def api_cat_facts():
    """Conecta a la API de Cat Facts y obtiene datos sobre gatos"""
    print("\n===== API 2: Cat Facts API =====")
    url = "https://cat-fact.herokuapp.com/facts"

    datos = conectar_api(url)
    if datos:
        print("Conexión exitosa con Cat Facts API!")
        print("\nEstructura JSON completa:")
        pprint(datos[:2])  # Mostramos solo los primeros 2 elementos

        print("\nExtracción de pares clave-valor:")
        # Mostramos información del primer hecho sobre gatos
        if datos and len(datos) > 0:
            primer_hecho = datos[0]
            print(f"Texto: {primer_hecho.get('text')}")
            print(f"Creado en: {primer_hecho.get('createdAt')}")
            print(f"Actualizado en: {primer_hecho.get('updatedAt')}")

            print("\nTodos los pares clave-valor del primer hecho:")
            extraer_pares_clave_valor(primer_hecho)
        else:
            print("No se encontraron hechos sobre gatos")
    else:
        print("No se pudieron obtener datos de la API de Cat Facts")

def api_open_brewery():
    """Conecta a la API de Open Brewery DB y obtiene datos sobre cervecerías"""
    print("\n===== API 3: Open Brewery DB =====")
    url = "https://api.openbrewerydb.org/breweries"
    params = {
        "by_city": "san_diego",
        "per_page": 3
    }

    datos = conectar_api(url, params=params)
    if datos:
        print("Conexión exitosa con Open Brewery DB!")
        print(f"Se encontraron {len(datos)} cervecerías en San Diego")

        print("\nEstructura JSON completa:")
        pprint(datos)

        print("\nExtracción de pares clave-valor de la primera cervecería:")
        if datos and len(datos) > 0:
            primera_cerveceria = datos[0]
            print(f"Nombre: {primera_cerveceria.get('name')}")
            print(f"Tipo: {primera_cerveceria.get('brewery_type')}")
            print(f"Dirección: {primera_cerveceria.get('street')}")
            print(f"Ciudad: {primera_cerveceria.get('city')}, {primera_cerveceria.get('state')}")
            print(f"Sitio web: {primera_cerveceria.get('website_url')}")

            print("\nTodos los pares clave-valor de la primera cervecería:")
            extraer_pares_clave_valor(primera_cerveceria)
        else:
            print("No se encontraron cervecerías")
    else:
        print("No se pudieron obtener datos de la API de Open Brewery DB")

def main():


    # Conexión a la API de Pokémon
    api_pokemon()

    # Conexión a la API de Cat Facts
    api_cat_facts()

    # Conexión a la API de Open Brewery DB
    api_open_brewery()



if __name__ == "__main__":
    main()
```
